```python
# ==================================== Read the Data File ==================================== #

data_raw = pd.read_csv("../10_data/SFM_data_byPlayer.csv")

# --- Some necessary Cosmetics
data_raw["goal"] = data_raw["goal"].astype(int)
data_raw["goals_in_match"] = data_raw["goals_in_match"].astype(int)
data_raw["kick_off"] = pd.to_datetime(
    data_raw["kick_off"], yearfirst=True
).dt.normalize()


# --- Compute share of games played on the left, right, center
data_raw["pos_share_LEFT"] = data_raw["N_games_left"] / data_raw[
    ["N_games_left", "N_games_center", "N_games_right"]
].sum(axis=1)
data_raw["pos_share_CENTER"] = data_raw["N_games_center"] / data_raw[
    ["N_games_left", "N_games_center", "N_games_right"]
].sum(axis=1)
data_raw["pos_share_RIGHT"] = data_raw["N_games_right"] / data_raw[
    ["N_games_left", "N_games_center", "N_games_right"]
].sum(axis=1)
positions = ["pos_share_LEFT", "pos_share_CENTER", "pos_share_RIGHT"]
# --- --- Unfortunately: drop NAs...shouldn't be too many
data_raw = data_raw.dropna(subset=positions).reset_index(drop=True)


# --- Final dataset to work with
data_raw = (
    data_raw[data_raw.name_player.isin(players)]
    .sort_values(["name_player", "kick_off"])
    .reset_index(drop=True)
)


# --- Indicator for Elite-Player
data_raw["elite_player"] = 0
data_raw.loc[data_raw["name_player"].isin(elite_players), "elite_player"] = 1


print(f"\nTotal Number of Observations: {data_raw.shape[0]}")
print(f'Total Number of Goals Scored: {data_raw["goals_in_match"].sum()}')


for g in [3, 4, 5]:
    g_elite = data_raw.loc[
        (data_raw["elite_player"] == 1) & (data_raw["goals_in_match"] == g), :
    ].shape[0]
    g_rlp = data_raw.loc[
        (data_raw["elite_player"] == 0) & (data_raw["goals_in_match"] == g), :
    ].shape[0]

    print(f"\nShare of Goals = {g} (Elite): {np.round(g_elite,3)}")
    print(f"Share of Goals = {g} (RLP):   {np.round(g_rlp,3)}")
```

### 1. &emsp; Frequency of Number of Goals

```python
# ============================== Frequency of Goals in Match by an Individual Player ============================== #

goals_elite = data_raw.loc[
    data_raw["elite_player"] == 1, "goals_in_match"
].value_counts()
goals_rlp = data_raw.loc[data_raw["elite_player"] == 0, "goals_in_match"].value_counts()

plot_df = pd.concat(
    [
        pd.DataFrame(
            {
                "N_goals": goals_elite.index,
                "total": goals_elite.values,
                "frequency_per_player": goals_elite.values
                / np.sum(data_raw["elite_player"] == 1),  # len(elite_players),
                "player_type": "Elite",
            }
        ),
        pd.DataFrame(
            {
                "N_goals": goals_rlp.index,
                "total": goals_rlp.values,
                "frequency_per_player": goals_rlp.values
                / np.sum(data_raw["elite_player"] == 0),  # len(rlp_players),
                "player_type": "RLP",
            }
        ),
    ]
)


# ------------------------------------------- Total Occurences of Number of Goals ------------------------------------------- #
fig, ax = plt.subplots(figsize=(7, 5))
sns.barplot(
    x=plot_df["N_goals"],
    y=plot_df["total"],
    hue=plot_df["player_type"],
    palette=[colors[7], colors[0]],
    ax=ax,
)
ax.legend(title="")
ax.set_xlabel("")
ax.set_ylabel("")
if export_fig:
    plt.savefig("../20_figures/Goals_Total.png")
plt.show()


# ------------------------------------------- Frequency of Number of Goals per Player ------------------------------------------- #
fig, ax = plt.subplots(figsize=(7, 5))
sns.barplot(
    x=plot_df["N_goals"],
    y=plot_df["frequency_per_player"],
    hue=plot_df["player_type"],
    palette=[colors[7], colors[0]],
    ax=ax,
)
ax.legend(title="")
ax.set_xlabel("")
ax.set_ylabel("")
if export_fig:
    plt.savefig("../20_figures/Goals_FrequencyByPlayerType.png")
plt.show()


# ------------------------------------------- Share of Goals > 3 ------------------------------------------- #

N3_elite = (
    plot_df.loc[
        (plot_df["player_type"] == "Elite") & (plot_df["N_goals"] > 3), "total"
    ].sum()
    / plot_df.loc[plot_df["player_type"] == "Elite", "total"].sum()
)
N3_rlp = (
    plot_df.loc[
        (plot_df["player_type"] == "RLP") & (plot_df["N_goals"] > 3), "total"
    ].sum()
    / plot_df.loc[plot_df["player_type"] == "RLP", "total"].sum()
)

print(f"\nShare of Goals > 3 (Elite): {np.round(N3_elite,3)}")
print(f"Share of Goals > 3 (RLP):   {np.round(N3_rlp,3)}")


for g in [3, 4, 5]:
    g_elite = (
        plot_df.loc[
            (plot_df["player_type"] == "Elite") & (plot_df["N_goals"] == g), "total"
        ].sum()
        / plot_df.loc[plot_df["player_type"] == "Elite", "total"].sum()
    )
    g_rlp = (
        plot_df.loc[
            (plot_df["player_type"] == "RLP") & (plot_df["N_goals"] == g), "total"
        ].sum()
        / plot_df.loc[plot_df["player_type"] == "RLP", "total"].sum()
    )

    print(f"\nShare of Goals = {g} (Elite): {np.round(g_elite,3)}")
    print(f"Share of Goals = {g} (RLP):   {np.round(g_rlp,3)}")
```

### 2. &emsp; Total Number of Goals by Season & League

```python
# ============================== Number of Goals in Match by Type of Player by League, and by Season ============================== #

# --- by League: Goals
goals_elite_league = (
    data_raw.loc[data_raw["elite_player"] == 1, :]
    .groupby("name_league")["goals_in_match"]
    .sum()
)
goals_rlp_league = (
    data_raw.loc[data_raw["elite_player"] == 0, :]
    .groupby("name_league")["goals_in_match"]
    .sum()
)
# --- by League: Players
players_elite_league = (
    data_raw.loc[data_raw["elite_player"] == 1, :]
    .groupby("name_league")["name_player"]
    .agg(lambda x: len(np.unique(x)))
)
players_rlp_league = (
    data_raw.loc[data_raw["elite_player"] == 0, :]
    .groupby("name_league")["name_player"]
    .agg(lambda x: len(np.unique(x)))
)
# --- by League: observations
observations_elite_league = (
    data_raw.loc[data_raw["elite_player"] == 1, :]
    .groupby("name_league")["name_player"]
    .agg(lambda x: len(x))
)
observations_rlp_league = (
    data_raw.loc[data_raw["elite_player"] == 0, :]
    .groupby("name_league")["name_player"]
    .agg(lambda x: len(x))
)


# --- by Season: Goals
goals_elite_season = (
    data_raw.loc[data_raw["elite_player"] == 1, :]
    .groupby("season")["goals_in_match"]
    .sum()
)
goals_rlp_season = (
    data_raw.loc[data_raw["elite_player"] == 0, :]
    .groupby("season")["goals_in_match"]
    .sum()
)


plot_df_league = pd.concat(
    [
        pd.DataFrame(
            {
                "name_league": [
                    " ".join([ll.capitalize() for ll in l.split("-")])
                    for l in goals_elite_league.index
                ],
                "total": goals_elite_league.values,
                "total_per_game": goals_elite_league.values
                / observations_elite_league.values,
                "player_type": "Elite",
            }
        ),
        pd.DataFrame(
            {
                "name_league": [
                    " ".join([ll.capitalize() for ll in l.split("-")])
                    for l in goals_rlp_league.index
                ],
                "total": goals_rlp_league.values,
                "total_per_game": goals_rlp_league.values
                / observations_rlp_league.values,
                "player_type": "RLP",
            }
        ),
    ]
)


plot_df_season = pd.concat(
    [
        pd.DataFrame(
            {
                "season": goals_elite_season.index,
                "total": goals_elite_season.values,
                "player_type": "Elite",
            }
        ),
        pd.DataFrame(
            {
                "season": goals_rlp_season.index,
                "total": goals_rlp_season.values,
                "player_type": "RLP",
            }
        ),
    ]
)


# ------------------------------------------- Number of Goals by League ------------------------------------------- #
fig, ax = plt.subplots(figsize=(9, 6.4))
sns.barplot(
    x=plot_df_league["name_league"],
    y=plot_df_league["total"],
    hue=plot_df_league["player_type"],
    palette=[colors[7], colors[0]],
    ax=ax,
)
ax.legend(title="")
ax.set_xlabel("")
ax.set_ylabel("")
if export_fig:
    plt.savefig("../20_figures/Goals_byLeague.png")
plt.show()


# ------------------------------------------- Number of Goals per Game by League ------------------------------------------- #
fig, ax = plt.subplots(figsize=(9, 6.4))
sns.barplot(
    x=plot_df_league["name_league"],
    y=plot_df_league["total_per_game"],
    hue=plot_df_league["player_type"],
    palette=[colors[7], colors[0]],
    ax=ax,
)
ax.legend(title="")
ax.set_xlabel("")
ax.set_ylabel("")
if export_fig:
    plt.savefig("../20_figures/Goals_PerGamebyLeague.png")
plt.show()


# ------------------------------------------- Number of Goals by Season ------------------------------------------- #
fig, ax = plt.subplots(figsize=(9, 6.4))
sns.barplot(
    x=plot_df_season["season"],
    y=plot_df_season["total"],
    hue=plot_df_season["player_type"],
    palette=[colors[7], colors[0]],
    ax=ax,
)
ax.legend(title="")
ax.set_xlabel("")
ax.set_ylabel("")
ax.set_xticks(np.arange(0, int(plot_df_season.shape[0] / 2), 2))
ax.tick_params(axis="x", labelrotation=90)
if export_fig:
    plt.savefig("../20_figures/Goals_bySeason.png")
plt.show()
```
